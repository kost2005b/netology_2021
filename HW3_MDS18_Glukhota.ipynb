# Задание 1

Дан объект в 2d пространстве:

```python
A = [[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]]

A = np.array(A)

x = A[:,0]
y = A[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

При помощи линейных отображений:
+ Уменьшить объект в два раза
+ Повернуть на 130 градусов
+ Отразить объект относительно прямой y=x
+ Вычислить определитель композиции преобразований (итоговой матрицы реализующей все три отображения

#### Уменьшаем

```python
#Линейный оператор для уменьшения (3 координату не трогаем)
B1  = np.array([[0.5, 0, 0], [0, 0.5, 0], [0, 0, 1]])
```

```python
A1  = (B1 @ A.T).T
A1
```

```python
A1 = np.array(A1)

x1 = A1[:,0]
y1 = A1[:,1]
plt.plot(x, y)

plt.plot(x1, y1, color = 'r')
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

#### Поворачиваем

```python
from math import cos, sin, pi
```

```python
#Переведем радианы в градусы
theta = 130 * 180/pi
```

```python
#Линейный оператор для поворота из лекции. С учетом того что поворачиваем по часовой стрелке
B2  = np.array([[ cos(theta), sin(theta), 0], [-sin(theta), cos(theta), 0], [0, 0, 1]])
```

```python
A2  = (B2 @ A1.T).T
```

```python
A2 = np.array(A2)

x2 = A2[:,0]
y2 = A2[:,1]

plt.plot(x, y)
plt.plot(x2, y2, color = 'g')
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

#### Отразим относительно y = x

```python
#Оператор для отражения по первым двум координатам
B3 = np.array([ [0,1,0], [1,0,0], [0,0,1] ])
```

```python
A3  = (B3 @ A2.T).T
```

```python
A3 = np.array(A3)

x3 = A3[:,0]
y3 = A3[:,1]

plt.plot(x, y)
plt.plot(x3, y3, color = 'y')

plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

#### Найдем определитель для композиции отображений

```python
#Важно умножить матрицы в той же последовательности, как  умножали выше. Для этого - обратный  порядок аргументов!
B_all = B3 @ B2 @ B1
```

```python
B_all
```

Убедимся что B_all дает нам то, что нужно

```python
A = [[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]]

A = np.array(A)
```

```python
A_all = (B_all @ A.T).T
```

```python
A_all = np.array(A_all)

x_all = A_all[:,0]
y_all = A_all[:,1]

x = A[:,0]
y = A[:,1]

plt.plot(x, y)
plt.plot(x_all, y_all, color = 'y')

plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

Рисунок тот же!

Найдем определитель для комбинации преобразований

```python
np.linalg.det(B_all)
```

# Задание 2

Решите систему уравнений, используя матричный подход

$x + z = 2$

$-x + y - 2z + w = -2$

$4x + z - 2w = 0$

$-4x + 4*y + w = 5$

Считаем вручную

```python
#Слева - к-ты, справа столбик со значениями
A = np.array([[1,0,1,0], [-1,1,-2,1], [4,0,1,-2], [-4,4,0,1]])
B = np.array([[2], [-2], [0], [5]])
```

```python
#Определитель находится - матрицу можно решать
np.linalg.det(A)
```

```python
#Находим обратную матрицу
A1 = np.linalg.inv(A)
```

```python
#Умножаем ее  на вектор B чтобы получить решение системы уравнений
A1 @  B
```

Исопльзуем встроенную функцию

```python
np.linalg.solve(A,B)
```

Так мы нашли значения всех 4 переменных
