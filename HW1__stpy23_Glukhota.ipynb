## Задание 2

Проанализировать значения по столбцам, рассчитать базовые статистики, найти выбросы.

```python
#посчитаем метрики для категориальных параметров

cat_list = ['surgery?', 'Age', 'pain', 'outcome', 'temperature of extremities']

for cat in cat_list:
    
    moda = df_[cat].mode()[0]
    unique_n = df_[cat].unique().size
    variants = df_[cat].unique()
    name_cat = str(cat)
    print(name_cat, ' - уникальных значений: ', unique_n, ' модальное значение: ', moda, 'значения: ', variants)
```

```python
#посчитаем метрики для числовых параметров

num_list = ['rectal temperature', 'pulse', 'respiratory rate']

for num in num_list:
    
    median_n = df_[num].median()
    mean_n = round(df_[num].mean(), 1)
    min_n = df_[num].min()
    max_n = df_[num].max()
    unique_n = df_[num].unique().size
    name_num = str(num)
    print(name_num, ' - уникальных значений: ', unique_n, ' медиана: ', median_n, 'среднее: ', mean_n, 'минимум: ', min_n,'максимум: ', max_n)
```

#### Скопируем датасет, чтобы заполнить пропуски

```python
import copy
```

```python
df_upd = copy.copy(df_)
```

#### Поле outcome и surgery

```python
#Попробуем заполнить модальным значением, за еимением лучшего варианта и подсказок
df_upd['outcome'].fillna(df_['outcome'].mode()[0], inplace = True)
df_upd['surgery?'].fillna(df_['surgery?'].mode()[0], inplace = True)
```

#### Поле pain

```python
#В расшифровках указано, что уровень боли связан с требуемой операцией - заполним модой по категориям surgery
df_upd['pain'].fillna(999, inplace = True)

for i in set(df_upd['surgery?'].tolist()):
    mode_i = df_upd[['surgery?', 'pain']][ df_upd['surgery?'] == i ].mode().loc[0][1]
    df_upd['pain'].loc[(df_upd['pain'] == 999) & (df_upd['surgery?'] == i)] = mode_i
```

#### Поле pulse

```python
#В расшифровках указано, что пульс может зависеть от боли у животного - попробуем посчитать медиану для категорий
df_upd['pulse'].fillna(df_upd.groupby('pain')['pulse'].transform('median'), inplace = True)
```

#### по rectal_temperature нет указаний для заполнения. Попробуем взять медиану, но это требует отдельного исследования

```python
df_upd['rectal_temperature'].fillna(df_['rectal_temperature'].median(), inplace = True)
```

#### temperature of extremities зависит от предыдущего (1 это  норма, но растет при высокой температуре)

```python
# можно проанализировать через квантили,  но не хватает экспертизы
#делаю оценку, что если температура выше нормы на 0.2, то warm (2), если ниже на 0.2 то cool (3), иначе normal (1)
df_upd['temperature_of_extremities'].fillna(999, inplace = True)

df_upd['temperature_of_extremities'].loc[(df_upd['temperature_of_extremities'] == 999) & (df_upd['rectal_temperature'] > 38.0)] = 2
df_upd['temperature_of_extremities'].loc[(df_upd['temperature_of_extremities'] == 999) & (df_upd['rectal_temperature'] < 37.6)] = 3
df_upd['temperature_of_extremities'].loc[df_upd['temperature_of_extremities'] == 999] = 3
```

#### зависимость respiratory rate от других показателей неочевидна. 

```python
#Есть некоторая корреляция с пульсом, но она тоже ненадежна. Мжно заполнить средним, но лучше вообще не использовать
df_upd['respiratory_rate'].fillna(df_['respiratory_rate'].mean(), inplace = True)
```
